# 7) Dessin

```python
# Un seul dessin sur le notebook

#créer une série temporelle
time=np.arange(0,100,0.1)
periode =10
signal = np.cos(2*np.pi*time/periode)
```

## Version courte, proche de matlab

```python

style=['-','.','o'] #je crée un tupple avec une liste de stype
plt.plot(time,signal,style[1]);
plt.xlabel('time sec')
plt.ylabel('amplitude')
plt.title('1er exemple')
plt.show() #inutile avec le notebook
```

## Version 2, avec subplot, proche de matlab

```python
# version 2
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(time,signal,style[0]);
plt.title('sinusoide')
plt.subplot(1,2,2)
plt.plot(time,np.sin(2*np.pi*time/periode),style[0]);
plt.title('co-sinusoide')
```

## Version plus logue avec des légendes

```python
#version longue avec des legendes
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(time,signal)
ax.set_xlabel('time sec')
ax.set_ylabel('amplitude')
ax.set_title('Une sinusoide')

#autre version longue avec des legendes
fig=plt.figure(figsize=(18,9))
ax=plt.subplot(1,1,1)
ax.plot(time,signal)
ax.set_xlabel('time sec')
ax.set_ylabel('amplitude')
ax.set_title('Une sinusoide');

# gca recupere le pointeur sur les axes
# gcf recupere le pointer sur la figure
# 
```
